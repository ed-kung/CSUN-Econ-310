```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm

SMALL_SIZE = 14
MEDIUM_SIZE = 16
LARGE_SIZE = 18
HUGE_SIZE = 20

plt.rcdefaults()
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = MEDIUM_SIZE
plt.rcParams['axes.titlesize'] = LARGE_SIZE
plt.rcParams['axes.labelsize'] = MEDIUM_SIZE
plt.rcParams['xtick.labelsize'] = MEDIUM_SIZE
plt.rcParams['ytick.labelsize'] = MEDIUM_SIZE
plt.rcParams['legend.fontsize'] = MEDIUM_SIZE
plt.rcParams['figure.titlesize'] = LARGE_SIZE
plt.rcParams['figure.figsize'] = [7.2, 7.2]
plt.rcParams['figure.dpi'] = 60
plt.rcParams['figure.facecolor'] = (1.0, 1.0, 1.0, 0.0)

from utils import SREQ, LREQ, CobbDouglasConsumer
```

```python
x = np.arange(0,11.1,0.1)
y = np.arange(0,11.1,0.1)
xgrid, ygrid = np.meshgrid(x,y)
ugrid = xgrid**0.5 * ygrid**0.5

fig, ax = plt.subplots()
ax.contour(xgrid,ygrid,ugrid, levels=np.arange(1,11), cmap=cm.coolwarm)
ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.arange(0,11))
ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_axisbelow(True)
plt.title(r'$u(x,y) = x^{\frac{1}{2}} y^{\frac{1}{2}}$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/12-exploring-utility-cobb-douglas.png', bbox_inches='tight')
plt.show()

```

```python
x = np.arange(0,11.1,0.1)
y = np.arange(0,11.1,0.1)
xgrid, ygrid = np.meshgrid(x,y)
ugrid = xgrid + ygrid

fig, ax = plt.subplots()
ax.contour(xgrid,ygrid,ugrid, levels=np.arange(2,21,2), cmap=cm.coolwarm)
ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.arange(0,11))
ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_axisbelow(True)
plt.title(r'$u(x,y) = x + y$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/12-exploring-utility-perfect-subs.png', bbox_inches='tight')
plt.show()

```

```python
x = np.arange(0,11.1,0.1)
y = np.arange(0,11.1,0.1)
xgrid, ygrid = np.meshgrid(x,y)
ugrid = np.minimum(xgrid, ygrid)

fig, ax = plt.subplots()
ax.contour(xgrid,ygrid,ugrid, levels=np.arange(1,11,1), cmap=cm.coolwarm)
ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.arange(0,11))
ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_axisbelow(True)
plt.title(r'$u(x,y) = \min(x,y)$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/12-exploring-utility-leontieff.png', bbox_inches='tight')
plt.show()

```

```python
x = np.arange(0,11.1,0.1)
y = np.arange(0,11.1,0.1)
xgrid, ygrid = np.meshgrid(x,y)
ugrid = xgrid + np.sqrt(ygrid)

fig, ax = plt.subplots()
ax.contour(xgrid,ygrid,ugrid, levels=np.arange(1,15,1), cmap=cm.coolwarm)
ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.arange(0,11))
ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_axisbelow(True)
plt.title(r'$u(x,y) = x + \sqrt{y}$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/12-exploring-utility-quasilinear.png', bbox_inches='tight')
plt.show()

```
