```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm

SMALL_SIZE = 14
MEDIUM_SIZE = 16
LARGE_SIZE = 18
HUGE_SIZE = 20

plt.rcdefaults()
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = MEDIUM_SIZE
plt.rcParams['axes.titlesize'] = LARGE_SIZE
plt.rcParams['axes.labelsize'] = MEDIUM_SIZE
plt.rcParams['xtick.labelsize'] = MEDIUM_SIZE
plt.rcParams['ytick.labelsize'] = MEDIUM_SIZE
plt.rcParams['legend.fontsize'] = MEDIUM_SIZE
plt.rcParams['figure.titlesize'] = LARGE_SIZE
plt.rcParams['figure.figsize'] = [7.2, 7.2]
plt.rcParams['figure.dpi'] = 60
plt.rcParams['figure.facecolor'] = (1.0, 1.0, 1.0, 0.0)

from utils import SREQ, LREQ, CobbDouglasConsumer
```

```python
x = np.arange(0,11.1,0.1)
y = np.arange(0,11.1,0.1)
xgrid, ygrid = np.meshgrid(x,y)
ugrid = xgrid**0.5 * ygrid**0.5

fig, ax = plt.subplots()
ax.contour(xgrid,ygrid,ugrid, levels=np.arange(1,11), cmap=cm.coolwarm)
ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.arange(0,11))
ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_axisbelow(True)
plt.title(r'$u(x,y) = x^{\frac{1}{2}} y^{\frac{1}{2}}$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/12-exploring-utility-cobb-douglas.png', bbox_inches='tight')
plt.show()
```

```python
a = 2/3
b = 1/3
px = 20
py = 10
I = 120
cb = CobbDouglasConsumer(a=a,b=b,px=px,py=py,I=I)
print(cb)

fig, ax = plt.subplots()
xmax = 12
ymax = 12
xticks = np.arange(0, xmax+1)
yticks = np.arange(0, ymax+1)
xgrid = np.arange(0.0, xmax+1.1, 0.1)
Ugrid = [cb.U]

for u in Ugrid:
    ax.plot(xgrid, (u/xgrid**a)**(1/b), color='green', alpha=0.3)

ax.plot(xgrid, (I-px*xgrid)/py, color='black')
ax.plot(cb.x, cb.y, 'o', color='black')

ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xlim(0,xmax+1)
ax.set_ylim(0,ymax+1)
ax.set_axisbelow(True)
plt.grid(alpha=0.2)
plt.savefig('../assets/images/12-exploring-utility-cb-example.png', bbox_inches='tight')
plt.show()
```
