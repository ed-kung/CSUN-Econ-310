```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm

SMALL_SIZE = 14
MEDIUM_SIZE = 16
LARGE_SIZE = 18
HUGE_SIZE = 20

plt.rcdefaults()
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = MEDIUM_SIZE
plt.rcParams['axes.titlesize'] = LARGE_SIZE
plt.rcParams['axes.labelsize'] = MEDIUM_SIZE
plt.rcParams['xtick.labelsize'] = MEDIUM_SIZE
plt.rcParams['ytick.labelsize'] = MEDIUM_SIZE
plt.rcParams['legend.fontsize'] = MEDIUM_SIZE
plt.rcParams['figure.titlesize'] = LARGE_SIZE
plt.rcParams['figure.figsize'] = [7.2, 7.2]
plt.rcParams['figure.dpi'] = 60
plt.rcParams['figure.facecolor'] = (1.0, 1.0, 1.0, 0.0)

from utils import SREQ, LREQ, CobbDouglasConsumer
```

```python
fig, ax = plt.subplots()

a = 0.25
b = 0.75
I = 6
px = 1
py = 1
v = 4

cb = CobbDouglasConsumer(a=a,b=b,I=I,px=px,py=py)
cb2 = CobbDouglasConsumer(a=a,b=b,I=I+v,px=px,py=py)


# Budget line (no food stamps)
xgrid = np.arange(0.1, 11, 0.1)
ybudg = (cb.I - cb.px*xgrid) / cb.py
ax.plot(xgrid, ybudg, color='black')

# Budget line (food stamps)
xgrid2 = np.arange(v, 11, 0.1)
ybudg2 = (cb.I+v - cb.px*xgrid2) / cb.py
ax.plot(xgrid2, ybudg2, color='blue')

# Budget line (voucher)
xgrid3 = np.arange(0.1, v, 0.1)
ybudg3 = (cb.I+v - cb.px*xgrid3) / cb.py
ax.plot(xgrid3, ybudg3, color='red', linestyle='dashed')

plt.legend(['Baseline', 'Food Stamps', 'Money Transfer'])

# Optimal point (no food stamps)
ax.plot(xgrid, (cb.U/xgrid**a)**(1/b), color='black')
ax.plot(cb.x, cb.y, 'o', color='black')
ax.text(cb.x+0.2, cb.y+0.2, 'A', color='black')

# Optimal point (food stamps)
U0 = v**a * (I/py)**b
ax.plot(xgrid, (U0/xgrid**a)**(1/b), color='blue')
ax.plot(v, I/py, 'o', color='blue')
ax.text(v-0.2, I/py-0.2, 'B', color='blue', horizontalalignment='right', verticalalignment='top')

# Optimal point (voucher)
ax.plot(xgrid, (cb2.U/xgrid**a)**(1/b), color='red')
ax.plot(cb2.x, cb2.y, 'o', color='red')
ax.text(cb2.x+0.2, cb2.y+0.2, 'C', color='red')

# Indifference curves
#xmesh, ymesh = np.meshgrid(np.arange(0.1,11,0.1), np.arange(0.1,11,0.1))
#zmesh = xmesh**a * ymesh**b
#CS = ax.contour(xmesh,ymesh,zmesh, levels=[cb.U, U0, cb2.U], cmap=cm.coolwarm)
#ax.clabel(CS, CS.levels, inline=True, fontsize=MEDIUM_SIZE)


ax.set_ylabel('Other Consumption')
ax.set_xlabel('Food Consumption')
ax.set_xticks(np.arange(0,11,1))
ax.set_yticks(np.arange(0,11,1))
ax.set_xticklabels(['']*11)
ax.set_yticklabels(['']*11)
ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_axisbelow(True)
plt.grid(alpha=0.2)
plt.title('Scenario B: Money transfer is better')
#plt.savefig('../assets/images/11-consumer-theory-perfect-substitutes.png', bbox_inches='tight')
plt.show()
```
