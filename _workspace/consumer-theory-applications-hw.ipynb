```python
import os
import numpy as np
import pandas as pd
from IPython.display import display, Markdown, Latex

import econtools
from econtools.documents import Homework, RawLatex
import utils2 as u2

from matplotlib import pyplot as plt

SMALL_SIZE = 14
MEDIUM_SIZE = 16
LARGE_SIZE = 18
HUGE_SIZE = 20

plt.rcdefaults()
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = MEDIUM_SIZE
plt.rcParams['axes.titlesize'] = LARGE_SIZE
plt.rcParams['axes.labelsize'] = MEDIUM_SIZE
plt.rcParams['xtick.labelsize'] = MEDIUM_SIZE
plt.rcParams['ytick.labelsize'] = MEDIUM_SIZE
plt.rcParams['legend.fontsize'] = MEDIUM_SIZE
plt.rcParams['figure.titlesize'] = LARGE_SIZE
plt.rcParams['figure.figsize'] = [7.2, 7.2]
plt.rcParams['figure.dpi'] = 60
plt.rcParams['figure.facecolor'] = (1.0, 1.0, 1.0, 0.0)

OUTPUT_DIR = "./tex"
IMAGE_PATH = "./tex"
TITLE = "Consumer Choice Theory Applications Problem Set"
PROBLEM_NAME = "consumer-theory-applications-hw-{}"
FILENAME = "consumer-choice-theory-applications-hw"

doc = Homework(
    title = TITLE,
)
rng = np.random.default_rng(4153)

nprob = 0
```

```python
# Price change problem: substitutes

TYPE = "PriceChangeProblem"
df = pd.read_csv(f"{TYPE}.csv")
params = dict(df.loc[df["comp_or_sub"]=="substitutes"].sample(1, random_state=rng).iloc[0])
name = fr"{FILENAME}-{nprob}"
problem = u2.load_problem(TYPE, params, rng=rng, name=name)
doc.add(u2.get_multipart_sa(problem,setup_id=0,question_ids=[0,1,2,3]))
problem.setup_axis.draw(saveas=f"{IMAGE_PATH}/{name}_setup.png")
plt.close()
nprob+=1
problem.sol
```

```python
# Price change problem: complements

TYPE = "PriceChangeProblem"
df = pd.read_csv(f"{TYPE}.csv")
params = dict(df.loc[df["comp_or_sub"]=="complements"].sample(1, random_state=rng).iloc[0])
name = fr"{FILENAME}-{nprob}"
problem = u2.load_problem(TYPE, params, rng=rng, name=name)
doc.add(u2.get_multipart_sa(problem,setup_id=0,question_ids=[0,1,2,3]))
problem.setup_axis.draw(saveas=f"{IMAGE_PATH}/{name}_setup.png")
plt.close()
nprob+=1
problem.sol
```

```python
# Public School: choose public, public more educ

TYPE = "PublicSchoolProblem"
df = pd.read_csv(f"{TYPE}.csv")
mask = (df['choice']=='public') & (df['x_public'] > df['x_private'])
params = dict(df.loc[mask].sample(1, random_state=rng).iloc[0])
name = fr"{FILENAME}-{nprob}"
problem = u2.load_problem(TYPE, params, rng=rng, name=name)
doc.add(u2.get_multipart_sa(problem,setup_id=0,question_ids=[0,1,2,3]))
problem.setup_axis.draw(saveas=f"{IMAGE_PATH}/{name}_setup.png")
plt.close()
nprob+=1
problem.sol
```
