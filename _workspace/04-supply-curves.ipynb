```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm

SMALL_SIZE = 12
MEDIUM_SIZE = 14
LARGE_SIZE = 16

plt.rcdefaults()
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = MEDIUM_SIZE
plt.rcParams['axes.titlesize'] = LARGE_SIZE
plt.rcParams['axes.labelsize'] = MEDIUM_SIZE
plt.rcParams['xtick.labelsize'] = SMALL_SIZE
plt.rcParams['ytick.labelsize'] = SMALL_SIZE
plt.rcParams['legend.fontsize'] = MEDIUM_SIZE
plt.rcParams['figure.titlesize'] = LARGE_SIZE
plt.rcParams['figure.figsize'] = [7.2, 4.8]
plt.rcParams['figure.dpi'] = 60
plt.rcParams['figure.facecolor'] = (1.0, 1.0, 1.0, 0.0)
```

```python
# Profit function
q = np.arange(0,10.1,0.1)
Pi = 10*q - q**2
plt.plot(q,Pi)
plt.xlabel(r'$q$')
plt.ylabel(r'$\Pi(q)$')
plt.xticks(np.arange(0,11,1))
plt.title(r'$\Pi(q) = 10q - q^2$')
plt.grid(alpha=0.2)
plt.xlim(0,10)
plt.ylim(0,30)
plt.savefig('../assets/images/04-supply-curves-profit-1.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# supply curve
q = np.arange(0,10.1,0.1)
p = 2*q
plt.plot(q,p)
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.xticks(np.arange(0,11,1))
plt.title(r'$q = p/2$')
plt.grid(alpha=0.2)
plt.xlim(0,10)
plt.ylim(0,20)
plt.savefig('../assets/images/04-supply-curves-supply-curve-1.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# producer surplus
q = np.arange(0,10.1,0.1)
p = 2*q
plt.plot(q,p)
plt.plot([0,10], [10,10], color='black')
plt.fill_between(q,p,10,where=(p<10),color='red',alpha=0.1)
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.xticks(np.arange(0,11,1))
plt.title(r'$q = p/2$')
plt.grid(alpha=0.2)
plt.xlim(0,10)
plt.ylim(0,20)
plt.savefig('../assets/images/04-supply-curves-producer-surplus-1.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# producer surplus
maxx = 2
stepx = 0.01
q = np.arange(0,maxx + stepx,stepx)
p = 3*q**2
myp = 5
myq = (myp/3)**0.5
myprofit = myp*myq - myq**3
plt.plot(q,p)
plt.plot([0,maxx], [myp,myp], color='black')
plt.fill_between(q,p,myp,where=(p<myp),color='red',alpha=0.1)
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.xticks(np.arange(0,11,1))
plt.title(r'$q = (p/3)^{1/2}$')
plt.grid(alpha=0.2)
plt.xlim(0,maxx)
plt.savefig('../assets/images/04-supply-curves-producer-surplus-2.png', transparent=False, bbox_inches="tight")
plt.show()

print(myq, myprofit)
```
