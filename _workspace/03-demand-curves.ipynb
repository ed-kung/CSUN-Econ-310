```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "font.size": 16,
    "figure.figsize": [7.2, 4.8],
    "figure.dpi": 60,
    "figure.facecolor": (1.0, 1.0, 1.0, 0.0)
})
```

```python
# Demand curve 1
q = np.arange(0,10.1,0.1)
p = 10 - q
plt.plot(q,p)
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.xticks(np.arange(0,11,1))
plt.yticks(np.arange(0,11,1))
plt.title(r'$q = 10 - p$')
plt.grid(alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,10)
plt.savefig('../assets/images/03-demand-curves-demand-curve-1.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# Consumer surplus
q = np.arange(0,10.1,0.1)
p = 10 - q
cs_points = [[0,10], [5,5], [0,5]]
cs_triangle = np.array(cs_points + cs_points[:1])
cs_triangle
plt.plot(q,p)
plt.plot([0,10],[5,5],color='black')
plt.fill(cs_triangle[:,0], cs_triangle[:,1], color='blue', alpha=0.1)
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.xticks(np.arange(0,11,1))
plt.yticks(np.arange(0,11,1))
plt.title(r'$q = 10 - p$')
plt.grid(alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,10)
plt.savefig('../assets/images/03-demand-curves-consumer-surplus-1.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# u(c,q) = c + bq^0.5,  Y=50
# p(q) = 0.5*b*q^-0.5
# q(p) = (2/b*p)^-2
# b = 2*p*q^0.5

Y = 50
b = 24

q = np.arange(0.1,20,0.1)
p = 0.5*b*q**-0.5

plt.plot(q,p)
plt.plot([0,20], [4,4], color='black')
plt.fill_between(q,p,4,where=(p>4),color='blue',alpha=0.1)
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.grid(alpha=0.2)
plt.ylim(0,15)
plt.xlim(0,20)
plt.savefig('../assets/images/03-demand-curves-consumer-surplus-2.png', transparent=False, bbox_inches="tight")
plt.show()

myp = 4
myq = (2/b*myp)**-2
myc = Y - myp*myq
myu = myc + b*myq**0.5

myp, myq, myc, myu
```
