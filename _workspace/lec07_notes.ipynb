```python
import os
import numpy as np
import pandas as pd
from IPython.display import display, Markdown, Latex
import json
import utils2 as u2
from matplotlib import pyplot as plt

SMALL_SIZE = 14
MEDIUM_SIZE = 16
LARGE_SIZE = 18
HUGE_SIZE = 20

plt.rcdefaults()
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = MEDIUM_SIZE
plt.rcParams['axes.titlesize'] = LARGE_SIZE
plt.rcParams['axes.labelsize'] = MEDIUM_SIZE
plt.rcParams['xtick.labelsize'] = MEDIUM_SIZE
plt.rcParams['ytick.labelsize'] = MEDIUM_SIZE
plt.rcParams['legend.fontsize'] = MEDIUM_SIZE
plt.rcParams['figure.titlesize'] = LARGE_SIZE
plt.rcParams['figure.figsize'] = [15, 7.2]
plt.rcParams['figure.dpi'] = 60
plt.rcParams['figure.facecolor'] = (1.0, 1.0, 1.0, 0.0)

IMAGE_PATH = "../assets/images/graphs"

rng = np.random.default_rng(101)
```

```python
xg = np.arange(0,13.1,0.1)
commodity_demand_1 = 12 - xg
commodity_supply_1 = 4 + xg
commodity_supply_2 = 0 + xg
labor_demand_1 = 8 - xg
labor_demand_2 = 12 - xg
labor_supply_1 = 0 + xg

plt.subplot(1,2,1)
plt.ylabel(r'$p$')
plt.xlabel(r'$q$')
plt.title('Commodity Market\n1st order effect of firm productivity shock')
plt.plot(xg, commodity_demand_1, color='blue')
plt.plot(xg, commodity_supply_1, color='red')
plt.plot(xg, commodity_supply_2, color='red', linestyle='dashed')
plt.arrow(6.25,10,3,0,color='red',head_width=0.3)
plt.xticks(np.arange(0,13))
plt.yticks(np.arange(0,13))
plt.xlim([0,13])
plt.ylim([0,13])
plt.grid(alpha=0.4)

plt.subplot(1,2,2)
plt.ylabel(r'$w$')
plt.xlabel(r'$L$')
plt.title('Labor Market\n1st order effect of firm productivity shock')
plt.plot(xg, labor_demand_1, color='blue')
plt.plot(xg, labor_demand_2, color='blue', linestyle='dashed')
plt.plot(xg, labor_supply_1, color='red')
plt.arrow(1.25,7,3,0,color='blue',head_width=0.3)
plt.xticks(np.arange(0,13))
plt.yticks(np.arange(0,13))
plt.xlim([0,13])
plt.ylim([0,13])
plt.grid(alpha=0.4)

plt.savefig(f"{IMAGE_PATH}/lec07_1st_order.png", bbox_inches='tight')
plt.show()
```

```python
xg = np.arange(0,13.1,0.1)
commodity_demand_1 = 12 - xg
commodity_supply_1 = 4 + xg
commodity_supply_2 = 0 + xg
commodity_supply_3 = 2 + xg
labor_demand_1 = 8 - xg
labor_demand_2 = 12 - xg
labor_demand_3 = 10 - xg
labor_supply_1 = 0 + xg

plt.subplot(1,2,1)
plt.ylabel(r'$p$')
plt.xlabel(r'$q$')
plt.title('Commodity Market\n2nd order effect of firm productivity shock')
plt.plot(xg, commodity_demand_1, color='blue')
plt.plot(xg, commodity_supply_1, color='red', alpha=0.1)
plt.plot(xg, commodity_supply_2, color='red', linestyle='dashed', alpha=0.3)
plt.plot(xg, commodity_supply_3, color='red', linestyle='dashdot', alpha=1.0)
plt.arrow(6.25,10,3,0,color='red',head_width=0.3,alpha=0.1)
plt.arrow(10.75,11,-1,0,color='red',head_width=0.3,alpha=1.0)
plt.xticks(np.arange(0,13))
plt.yticks(np.arange(0,13))
plt.xlim([0,13])
plt.ylim([0,13])
plt.grid(alpha=0.4)

plt.subplot(1,2,2)
plt.ylabel(r'$w$')
plt.xlabel(r'$L$')
plt.title('Labor Market\n2nd order effect of firm productivity shock')
plt.plot(xg, labor_demand_1, color='blue', alpha=0.1)
plt.plot(xg, labor_demand_2, color='blue', linestyle='dashed', alpha=0.3)
plt.plot(xg, labor_demand_3, color='blue', linestyle='solid', alpha=1.0)
plt.plot(xg, labor_supply_1, color='red')
plt.arrow(1.25,7,3,0,color='blue',head_width=0.3, alpha=0.1)
plt.arrow(3.75,8,-1,0,color='blue',head_width=0.3, alpha=1)
plt.xticks(np.arange(0,13))
plt.yticks(np.arange(0,13))
plt.xlim([0,13])
plt.ylim([0,13])
plt.grid(alpha=0.4)

plt.savefig(f"{IMAGE_PATH}/lec07_2nd_order.png", bbox_inches='tight')
plt.show()
```
