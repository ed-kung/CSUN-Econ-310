```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "font.size": 12
})
```

```python
# f(x) = x^2
x = np.arange(0,1,0.01)
y = x**2
plt.plot(x,y)
plt.title(r'$f(x)=x^2$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/01-math-review-xsquared.png', transparent=False)
plt.show()
```

```python
# f(x) = sqrt(x)
x = np.arange(0,1,0.01)
y = np.sqrt(x)
plt.plot(x,y)
plt.title(r'$f(x)=\sqrt{x}$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/01-math-review-sqrtx.png', transparent=False)
plt.show()
```

```python
# f(x,y) = x^0.5 y^0.5
x = np.arange(0,1,0.01)
y = np.arange(0,1,0.01)
x, y = np.meshgrid(x, y)
z = x**0.5 * y**0.5
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})

ax.plot_surface(x,y,z,cmap=cm.coolwarm)
plt.title(r'$f(x,y)=x^{0.5}y^{0.5}$')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$f(x,y)$')
plt.savefig('../assets/images/01-math-review-sqrtxy.png', transparent=False)
plt.show()
```

```python
# Supply and demand
q = np.arange(0,10,0.1)
pd = 10 - q
ps = q
plt.plot(q, pd)
plt.plot(q, ps)
plt.legend(['Demand','Supply'])
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.grid(alpha=0.2)
plt.savefig('../assets/images/01-math-review-supply-demand-01.png')
plt.show()
```

```python
# Supply and demand 2
q = np.arange(0.1, 10, 0.1)
pd = 10/q
ps = (1/9)*q**2
plt.plot(q, pd)
plt.plot(q, ps)
plt.legend(['Demand','Supply'])
plt.xlabel(r'$q$')
plt.ylabel(r'$p$')
plt.grid(alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,10)
plt.savefig('../assets/images/01-math-review-supply-demand-02.png')
plt.show()
p = (10/3)**(2/3)
print(p)
q = 10/p
print(q)
```
