```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "font.size": 12
})
```

```python
x = np.arange(0,2,0.01)
y1 = x**0.25
y2 = x**0.5
y3 = x**1
y4 = x**2
y5 = x**4
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.xlim(0,2)
plt.ylim(0,2)
plt.grid(alpha=0.2)
plt.title(r'$y = x^b$')
plt.legend([
    r'$b=1/4$',
    r'$b=1/2$',
    r'$b=1$',
    r'$b=2$',
    r'$b=4$'
])
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.savefig('../assets/images/01-math-review-polynomial-graph.png')
plt.show()
```
