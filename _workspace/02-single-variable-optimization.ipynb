```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "font.size": 16,
    "figure.figsize": [7.2, 4.8],
    "figure.dpi": 60,
    "figure.facecolor": (1.0, 1.0, 1.0, 0.0)
})
```

```python
# Graphical intuition
x = np.arange(0,5.21,0.1)
y = x*(5.2-x)
plt.plot(x,y)
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.title(r'$f(x)$')
plt.grid(alpha=0.2)
plt.ylim(0,9)
plt.xlim(0,5.2)
plt.savefig('../assets/images/02-single-variable-optimization-intuition.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# Calculating slope
x = np.arange(0,5.21,0.1)
y = x*(5.2-x)
myx = x[20]
myy = y[20]
myslope = 5.2 - 2*myx
x2 = np.arange(myx-1.5, 3.8, 0.1)
y2 = myy + myslope*(x2 - myx)
plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(myx,myy,'o',color='black')

# dy / dx
myx1 = 2.8
myx2 = 3.6
myy1 = myy + myslope*(myx1 - myx)
myy2 = myy + myslope*(myx2 - myx)
plt.plot([myx1, myx2], [myy1, myy1], color='black', linestyle='dashed')
plt.plot([myx2, myx2], [myy1, myy2], color='black', linestyle='dashed')
plt.text((myx1+myx2)/2, myy1-0.5, r'$\Delta x$', horizontalalignment='center')
plt.text(myx2+0.075, (myy1+myy2)/2, r'$\Delta y$', verticalalignment='center')
plt.text(myx-0.1, myy-0.8,r'Slope = $\Delta y / \Delta x$')

plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid(alpha=0.2)
plt.ylim(0,9)
plt.xlim(0,5.2)
plt.savefig('../assets/images/02-single-variable-optimization-slope.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# Slope is zero at maximum
x = np.arange(0,5.21,0.1)
y = x*(5.2-x)
myx = 2.6
myy = myx*(5.2-myx)
myslope = 5.2 - 2*myx
x2 = np.arange(myx-1.5, 3.8, 0.1)
y2 = myy + myslope*(x2 - myx)
plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(myx,myy,'o',color='black')

# dy / dx
plt.text(myx, myy-0.8,r'Slope = 0',horizontalalignment='center')

plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid(alpha=0.2)
plt.ylim(0,9)
plt.xlim(0,5.2)
plt.savefig('../assets/images/02-single-variable-optimization-slope-at-max.png', transparent=False, bbox_inches="tight")
plt.show()
```

```python
# Graphical intuition
x = np.arange(0,5.21,0.1)
y = 5.2 - 2*x
plt.plot([0,5.2],[0,0],color='black')
plt.plot(x,y)
plt.title(r'$f^\prime(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f^\prime(x)$')
plt.grid(alpha=0.2)
plt.xlim(0,5.2)
plt.savefig('../assets/images/02-single-variable-optimization-derivative.png', transparent=False, bbox_inches="tight")
plt.show()
```
