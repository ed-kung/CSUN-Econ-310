```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm

SMALL_SIZE = 14
MEDIUM_SIZE = 16
LARGE_SIZE = 18
HUGE_SIZE = 20

plt.rcdefaults()
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = MEDIUM_SIZE
plt.rcParams['axes.titlesize'] = LARGE_SIZE
plt.rcParams['axes.labelsize'] = MEDIUM_SIZE
plt.rcParams['xtick.labelsize'] = MEDIUM_SIZE
plt.rcParams['ytick.labelsize'] = MEDIUM_SIZE
plt.rcParams['legend.fontsize'] = MEDIUM_SIZE
plt.rcParams['figure.titlesize'] = LARGE_SIZE
plt.rcParams['figure.figsize'] = [7.2, 7.2]
plt.rcParams['figure.dpi'] = 60
plt.rcParams['figure.facecolor'] = (1.0, 1.0, 1.0, 0.0)

from utils import SREQ, LREQ
```

```python
x = np.arange(0,11.1,0.1)
y = np.arange(0,11.1,0.1)
xgrid, ygrid = np.meshgrid(x,y)
ugrid = xgrid**0.5 * ygrid**0.5
budget_y = 10 - x

fig, ax = plt.subplots()
ax.contour(xgrid,ygrid,ugrid, levels=np.arange(2,10), cmap=cm.winter)
#ax.clabel(CS, CS.levels, inline=True, fontsize=MEDIUM_SIZE)
ax.plot(x,budget_y,color='black',linewidth=3)
ax.plot(5,5,'o',color='black',markersize=14)
ax.text(5.2,5.2,'Optimal choice')
ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.arange(0,11))
ax.set_xticklabels(['']*10 + [r'$\frac{I}{p_x}$'], fontsize=HUGE_SIZE)
ax.set_yticklabels(['']*10 + [r'$\frac{I}{p_y}$'], fontsize=HUGE_SIZE)
ax.set_xlim(0,11)
ax.set_ylim(0,11)
ax.set_axisbelow(True)
plt.grid(alpha=0.2)
plt.savefig('../assets/images/11-consumer-theory-graph-1.png', bbox_inches='tight')
plt.show()

```
