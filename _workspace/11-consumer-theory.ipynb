## Cobb Douglas consumer problem

$$ \max_{x,y} ~  u(x,y) = x^{a} y^{b} ~ ~ \text{ s.t. } ~ ~ p_x x + p_y y = I $$

$$ x = \frac{I}{p_x(1+b/a)} $$
$$ y = \frac{I}{p_y(1+a/b)} $$

Iso-utility curves:

$$ y = \left( \frac{U}{x^a} \right)^{1/b} $$

Budget line:

$$ y = \frac{I - p_x x}{p_y} $$


```python
a = 0.5
b = 0.5
I = 100
px = 5
py = 10

cb = CobbDouglasConsumer(a=a,b=b,I=I,px=px,py=py)

assert cb.is_integer()

print(cb)

xgrid = np.arange(0.1, 21, 0.1)
ybudg = (cb.I - cb.px*xgrid) / cb.py
yiso = (cb.U / xgrid**cb.a )**(1/cb.b)

fig, ax = plt.subplots()
ax.plot(xgrid, ybudg)
ax.plot(xgrid, yiso)
ax.plot(cb.x, cb.y, 'o', color='black')
ax.set_ylabel(r'$y$')
ax.set_xlabel(r'$x$')
ax.set_xticks(np.arange(0,21,2))
ax.set_yticks(np.arange(0,21,2))
ax.set_xlim(0,21)
ax.set_ylim(0,21)
ax.set_axisbelow(True)
plt.grid(alpha=0.2)
plt.savefig('../assets/images/11-consumer-theory-example-5.png', bbox_inches='tight')
plt.show()
```
