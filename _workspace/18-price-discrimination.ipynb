```python
# Consumer surplus
q = np.arange(0,10.1,0.1)

# Perfect competition
plt.subplot(1,3,1)
pd = 10 - q
ps = q
cs_points = [[0,10], [5,5], [0,5]]
cs_triangle = np.array(cs_points + cs_points[:1])
ps_points = [[0,0], [5,5], [0,5]]
ps_triangle = np.array(ps_points + ps_points[:1])
plt.plot(q,pd,color='blue')
plt.plot(q,ps,color='red')
plt.plot([0,5],[5,5],color='black',linestyle='dashed')
plt.plot([5,5],[0,5],color='black',linestyle='dashed')
plt.fill(cs_triangle[:,0], cs_triangle[:,1], color='blue', alpha=0.1)
plt.fill(ps_triangle[:,0], ps_triangle[:,1], color='red', alpha=0.1)
plt.text(-0.2,5,r'$p$',horizontalalignment='right',verticalalignment='center')
plt.text(5,-0.2,r'$q$',verticalalignment='top',horizontalalignment='center')
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.yticks([])
plt.title('Surplus in Perfect Competition')
plt.grid(alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,10)

# Monopoly
myp = 6.66667
myq = 3.33333
plt.subplot(1,3,2)
pd = 10 - q
mr = 10 - 2*np.arange(0,5.1,0.1)
ps = q
cs_points = [[0,10], [myq,myp], [0,myp]]
cs_triangle = np.array(cs_points + cs_points[:1])
ps_points = [[0,0], [myq,myq], [myq,myp], [0,myp]]
ps_triangle = np.array(ps_points + ps_points[:1])
plt.plot(q,pd,color='blue')
plt.plot(np.arange(0,5.1,0.1), mr, color='blue', linestyle='dashed')
plt.plot(q,ps,color='red')
plt.plot([0,myq],[myp,myp],color='black',linestyle='dashed')
plt.plot([myq,myq],[0,myp],color='black',linestyle='dashed')
plt.fill(cs_triangle[:,0], cs_triangle[:,1], color='blue', alpha=0.1)
plt.fill(ps_triangle[:,0], ps_triangle[:,1], color='red', alpha=0.1)
plt.text(-0.2,myp,r'$p$',horizontalalignment='right',verticalalignment='center')
plt.text(myq,-0.2,r'$q$',verticalalignment='top',horizontalalignment='center')
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.yticks([])
plt.title('Surplus in Monopoly')
plt.grid(alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,10)

# Perfect price discrimination
plt.subplot(1,3,3)
pd = 10 - q
ps = q
ps_points = [[0,0], [5,5], [0,10]]
ps_triangle = np.array(ps_points + ps_points[:1])
plt.plot(q,pd,color='blue')
plt.plot(q,ps,color='red')
plt.fill(ps_triangle[:,0], ps_triangle[:,1], color='red', alpha=0.1)
plt.plot([5,5],[0,5],color='black',linestyle='dashed')
plt.text(5,-0.2,r'$q$',verticalalignment='top',horizontalalignment='center')
plt.xlabel('')
plt.ylabel('')
plt.xticks([])
plt.yticks([])
plt.title('Surplus in Perfect Price Discrimination')
plt.grid(alpha=0.2)
plt.ylim(0,10)
plt.xlim(0,10)

plt.savefig('../assets/images/18-price-discrimination-1st-degree.png', transparent=False, bbox_inches="tight")
plt.show()
```
