```python
examples = []
n_found = 0
for w1 in np.arange(15,105,5):
    for w2 in np.arange(15,105,5):
        for cmax in np.array([60,90,120])*max(w1,w2):
            for d in np.arange(2,5):
                for nx in np.arange(1,d):
                    ny = d - nx
                    params = {'nx':nx,'dx':d,'ny':ny,'dy':d,'w1':w1,'w2':w2,'cmax':cmax}
                    prob = WageChangeCB(params)
                    if prob.check_solution() and (w1!=w2):
                        n_found+=1
                        examples.append( prob.params | prob.sol )
                        if (n_found%1000==0):
                            print(f"{n_found} examples found...")
examples = pd.DataFrame.from_dict(examples)
examples = examples.drop_duplicates()
examples.to_csv("WageChangeCB_examples.csv", index=False, header=True)
print(f"{len(examples)} examples found.")
```
